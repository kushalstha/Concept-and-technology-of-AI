Problem - 4: Matrix Operations:
• For the following arrays:
A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]);
Prove following with Numpy:

1. Prove A.A−1 = I.

```python
import numpy as np
A = np.array([[3,4],
              [7,8]])
B = np.array([[5,3],
            [2,1]])

A_inv = np.linalg.inv(A)
I = np.dot(A, A_inv)
print("A * A^-1: \n",I)
```

2. Prove AB ̸= BA.


```python
import numpy as np
A = np.array([[3,4],
              [7,8]])
B = np.array([[5,3],
            [2,1]])

AB = np.dot(A, B)
print("AB:\n ",AB)
BA = np.dot(B, A)
print("BA:\n ",BA)

print("AB == BA?: \n", np.array_equal(AB, BA))
```

3. Prove (AB)^T = B^TA^T

```python
import numpy as np
A = np.array([[3,4],
              [7,8]])
B = np.array([[5,3],
            [2,1]])

AB = np.dot(A, B)
AB_T = AB.T
print("(AB)^T: \n",AB_T)

BT_AT = np.dot(B.T, A.T)
print("B^T . A^T: \n", BT_AT)

print("(AB)^T == B^T . A^T?: \n", np.array_equal(AB_T, BT_AT))
```

Solve the following system of Linear equation using Inverse Methods.

2x − 3y + z = −1

x − y + 2z = −3

3x + y − z = 9

{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}

```python
import numpy as np
A = np.array([[2, -3, 1],
              [1, -1, 2],
              [3, 1, -1]])

B = np.array([-1, -3, 9])
K = np.linalg.solve(A, B)
print("Solution of [x,y,z]: \n",K)
```

Now: solve the above equation using np.linalg.inv function.{Explore more about ”linalg” function
of Numpy}

```python
import numpy as np
A = np.array([[2, -3, 1],
              [1, -1, 2],
              [3, 1, -1]])

B = np.array([-1, -3, 9])

A_inv =  np.linalg.inv(A)
X = np.dot(A_inv, B)
print("Solution of [x, y, z]: \n",X)
```

4. Matrix Multiplication
• Using Python lists, perform matrix multiplication of two matrices of size 1000x1000. Measure
and print the time taken for this operation.

```python
import numpy as np
import time
size = 1000

mat1 = [[i + j for j in range(size)] for i in range(size)]
mat2 = [[i + j for j in range(size)] for i in range(size)]

start = time.time()

result = [[0]* size for _ in range(size)]

for i in range(size):
  for j in range(size):
    for k in range(size):
      result[i][j] += mat1[i][k] * mat2[k][j]

end = time.time()
print("python list matrix multiplication time: ", end - start , "seconds")
```

• Using NumPy arrays, perform matrix multiplication of two matrices of size 1000x1000. Mea-

```python
import numpy as np
import time
size = 1000

mat1_np = np.arange(size*size).reshape(size, size)
mat2_np = np.arange(size*size).reshape(size, size)

start = time.time()
result_np = np.dot(mat1_np, mat2_np)
end = time.time()

print("NumPy matrix multiplication time:", end - start, "seconds")
```
